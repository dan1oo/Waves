```python
lengths = [0.400, 0.600, 0.800, 1.000, 1.200, 1.400, 1.600]
freq = [391.77, 261.73, 197.91, 157.17, 129.21, 110.97, 102.13]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(L,a):
    return a/L

popt, pcov = curve_fit(func, lengths, freq)
```

```python
a = popt[0]

print("The best fit parameters are:")
print("a = " + str(a))
print("f(L) = "+str(a)+"/L")
perr = np.sqrt(np.diag(pcov))
print("Uncertainty: "+ str(perr[0]))
```

```python
xvalues = np.linspace(0.4,1.6,100)
plt.figure(2)
plt.plot(lengths, freq,'o', color = "black",markersize = 4, label = "data")
plt.plot(xvalues, func(xvalues,a), color = "black", linewidth = 1.6, label = "best fit model")
plt.xlabel("Length",fontsize = 14)
plt.ylabel("f(L)",fontsize = 14)
plt.legend()
plt.show()
```
